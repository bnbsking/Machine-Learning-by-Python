```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt, pi, log
```

```python
def gauss(mean, sigma, X):
    Y = 1/(sigma*sqrt(2*pi))*np.exp((-1/(2*sigma**2))*(X-mean)**2)
    return Y/np.sum(Y)
    
def info_entropy(p):
    return -np.sum(p*np.log(p))
```

```python
# continuous shannon entropy
X = np.linspace(0,1,10000)
Y1 = [0.0001]*10000
Y2 = gauss(0.5, 0.1, X)
Y3 = gauss(0.5, 0.05, X) 
plt.subplot(2,1,1)
plt.title('Probability distributions')
plt.plot(X, Y1); plt.plot(X, Y2); plt.plot(X, Y3)
plt.legend(labels=['constant', 'blunt_gauss', 'sharp_gauss'],  loc='best')

E1, E2, E3 = info_entropy(Y1), info_entropy(Y2), info_entropy(Y3)
plt.subplot(2,1,2)
plt.title('Information entropy')
plt.bar(['constant', 'blunt_gauss', 'sharp_gauss'], [E1, E2, E3])
plt.show()
```
